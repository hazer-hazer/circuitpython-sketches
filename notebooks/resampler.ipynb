```python
from array import array
from random import randint
from math import sin, pi
from time import time, time_ns
import matplotlib.pyplot as plt
from lib.dsp.resampler import Resampler
from lib.dsp.fir_lpf import FirLpf

sample_rate = 44100
pitch = 440
length = int(sample_rate / pitch)
sine_wave = array("H", [0] * length)
for i in range(length):
    sine_wave[i] = int(sin(pi * 2 * i / length) * ((2**15) - 1) + 2**15)

resampled_rate = sample_rate // 2
resampled = Resampler().resample(sine_wave, from_rate=sample_rate, to_rate=resampled_rate, use_lpf=None)

for x, y in enumerate(sine_wave):
    plt.plot(x, y, 'g.', label='Sine wave')

for x, y in enumerate(resampled):
    plt.plot(x, y, 'b.', label='Resampled')

plt.show()
```
